```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 12 16:11:07 2024

@author: lucas
"""

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Connect to SQLite database
con = sqlite3.connect("C:/Users/lucas/Desktop/Data Science/Chessdotcom/chess_analytics.db") 

# Query the database and store the result in a dataframe
query = "SELECT * FROM opponent_vs_user_distribution"
distribution = pd.read_sql_query(query, con)
```

```python
# color palette plus standards
colors = ['#118DFF', '#573B92', '#ffa600', '#E91E63', '#9C27B0', '#d43d51']

plt.figure(frameon=False)

# histogram of the user
plt.hist(distribution['average_user_rating'], bins=25, density=True, color=colors[2], alpha=0.7, edgecolor='black', linewidth=1.2)

# normal distribution density curve
mu_opponent = np.mean(distribution['average_user_rating'])
sigma_opponent = np.std(distribution['average_user_rating'])
x_opponent = np.linspace(min(distribution['average_user_rating']), max(distribution['average_user_rating']), 100)
plt.plot(x_opponent, 1/(sigma_opponent * np.sqrt(2 * np.pi)) * np.exp(- (x_opponent - mu_opponent)**2 / (2 * sigma_opponent**2)), color=colors[5])

plt.xlabel('User rating', fontname='Segoe UI')
plt.ylabel('Density', fontname='Segoe UI')
plt.xticks(fontname='Segoe UI')
plt.yticks(fontname='Segoe UI')
plt.grid(False)

for spine in plt.gca().spines.values():
    spine.set_visible(False)


plt.show()
```
