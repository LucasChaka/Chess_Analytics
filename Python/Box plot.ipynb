```python
# Paste or type your script code here:
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Connect to SQLite database
con = sqlite3.connect("C:/Users/lucas/Desktop/Data Science/Chessdotcom/chess_analytics.db") 

# Query the database and store the result in a dataframe
query = "SELECT * FROM opponent_vs_user_distribution"
distribution = pd.read_sql_query(query, con)

# 

colors = ['#0c6185', '#00876c', '#F5A76D', '#E91E63', '#9C27B0', '#d43d51']


plt.boxplot(distribution['average_user_rating'], patch_artist=True, boxprops=dict(facecolor=colors[1], edgecolor=colors[1]), whiskerprops=dict(color='black'), capprops=dict(color='black'), medianprops=dict(color='black'), flierprops=dict(marker='o', markerfacecolor=colors[2], markersize=8))

plt.ylabel('User rating', fontname='Segoe UI')
plt.xticks(fontname='Segoe UI')
plt.yticks(fontname='Segoe UI')
plt.grid(False)

# Hide spines
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Show the plot
plt.show()
```
